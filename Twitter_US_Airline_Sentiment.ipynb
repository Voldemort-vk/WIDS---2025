```python
import pandas as pd

Tweet_dataset = pd.read_csv("WIDS---2025/Tweets.csv", encoding="latin-1")
```

```python
Tweet_dataset.head()
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.lower() #LOWER CASE
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"@\w+", "", regex = True) #REMOVE MENTIONS
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"#\w+", "", regex = True) #REMOVE HASHTAGS
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"(\w+)n't", r"\1 not", regex = True)
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"\bwon't\b", r"will not", regex = True)
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"\bit's\b", r"it is", regex = True)
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(r"(\w+)'ve", r"\1 have", regex = True)
```

```python
import string
Tweet_dataset['text'] = Tweet_dataset['text'].str.replace(f"[{string.punctuation}]", "", regex = True) #REMOVE PUNCTUATION
```

```python
Tweet_dataset.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatizing_tweets(tweet):
  tokens = word_tokenize(tweet)
  lemmatized_tokens = []
  for token in tokens:
    lemmatized_token = lemmatizer.lemmatize(token)
    lemmatized_tokens.append(lemmatized_token)

  return ' '.join(lemmatized_tokens)
```

```python
Tweet_dataset['text'] = Tweet_dataset['text'].apply(lemmatizing_tweets)
```

```python
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format(word2vec_path, binary=True)
```

```python
import numpy as np
def vectorize_tweets(tweet, model, vector_size = 300): #vector_length = 300 fo google news
  tokens = word_tokenize(tweet)
  token_vectors = []
  for token in tokens:
    if token in model:
      token_vector = model[token]
      token_vectors.append(token_vector)

  if len(token_vectors) == 0:
    return np.zeros(vector_size)
  else:
    return np.mean(token_vectors, axis = 0)
```

```python
Tweet_vectors = []
for tweet in Tweet_dataset['text']:
  tweet_vector = vectorize_tweets(tweet, model)
  Tweet_vectors.append(tweet_vector)

Tweet_vectors = np.array(Tweet_vectors)
```

```python
Tweet_vectors.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()

Data = Tweet_vectors
Labels = encoder.fit_transform(Tweet_dataset['airline_sentiment'])

Data_train, Data_test, Labels_train, Labels_test = train_test_split(Data, Labels, test_size = 0.2, shuffle = True)
```

```python
encoder.classes_
```

```python
print(Data_train.shape, Data_test.shape)
print(Labels_train.shape, Labels_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

MLR = LogisticRegression(multi_class='multinomial', solver='lbfgs')
MLR.fit(Data_train, Labels_train)

Predicted_Labels = MLR.predict(Data_test)
print("Test Accuracy:", accuracy_score(Labels_test, Predicted_Labels))
```

```python
import re

def predict_tweet_sentiment(trained_model, word2vec_model, tweet):
  encoded_array = np.array(['Negative', 'Neutral', 'Positive']) #After checking the encoding of the encoder used above
  tweet = tweet.lower()
  tweet = re.sub(r'@\w+', '', tweet)
  tweet = re.sub(r'#\w+', '', tweet)
  tweet = re.sub(r"(\w+)n't", r"\1 not", tweet)
  tweet = re.sub(r"\bwon't\b", "will not", tweet)
  tweet = re.sub(r"\bit's\b", "it is", tweet)
  tweet = re.sub(f"[{string.punctuation}]", "", tweet)

  tweet_vector = vectorize_tweets(tweet, word2vec_model)
  tweet_vector = tweet_vector.reshape(1, -1)

  predicted_label = trained_model.predict(tweet_vector)[0]

  sentiment = encoded_array[predicted_label]

  return sentiment
```

```python
#Examples
tweet1 = "Very Bad Flight!!"
tweet2 = "Flight was nice:)"
tweet3 = ""

sentiment1 = predict_tweet_sentiment(MLR, model, tweet1)
sentiment2 = predict_tweet_sentiment(MLR, model, tweet2)
sentiment3 = predict_tweet_sentiment(MLR, model, tweet3)

print(f"Tweet: {tweet1}, Sentiment: {sentiment1}")
print(f"Tweet: {tweet2}, Sentiment: {sentiment2}")
print(f"Tweet: {tweet3}, Sentiment: {sentiment3}")
```
